```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib 
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
df1 = pd.read_csv("Bengaluru_House_Data.csv")
df1.head()
```

```python
df1.groupby('area_type')['area_type'].agg('count')
```

```python
#here we are going to drop Some unused colums--

df2 = df1.drop(['area_type', 'society', 'balcony', 'availability'], axis = 'columns')
df2.head()
```

```python
#Now we are going to clean the data - 
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.shape
```

```python
df3['size'].unique()
```

```python
#here we are going add a new col - name - bhk - which is going to have data in numbers only.
df3['bhk'] = df3['size'].apply(lambda x:int(x.split(' ')[0]))
```

```python
df3['bhk'].unique()
```

```python
df3[df3.bhk>20]
```

```python
df3.total_sqft.unique()
```

```python
#here we are see some value like 1133 - 1384  where the data is not fix we are going to change it
# this funtion is going to chage the value in float and if it is not a valid numbere it is going to delet it
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
    
```

```python
df3[~df3['total_sqft'].apply(is_float)].head(10)
```

```python
df3.head()
```

```python
#here we are see some value like 1133 - 1384  where the data is not fix we are going to change it
# this funtion is going to chage the value in float and if it is not a valid numbere it is going to chage it
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
        
```

```python
convert_sqft_to_num('2020-2202')
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
df4.head(4)
```

```python
#here we are just searching an element from its locations 
df4.loc[30]
```

```python
df5 = df4.copy()
```

```python
#here we are creating a new feature - 
df5['price_per_sqft'] = df5['price'] * 100000/df5['total_sqft']
df5.head()
```

```python
len(df5.location.unique())
```

```python
#now we are removing the dimension or location for future - 
# here we are making another 
#any location which has less the 10 data is in the other 
df5.location = df5.location.apply(lambda x: x.strip())

location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending = False)
location_stats               
```

```python
#any location which has less the 10 data is in other 
len(location_stats[location_stats<=10])
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
location_stats_less_than_10
```

```python
df5.location = df5.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
len(df5.location.unique())
```
